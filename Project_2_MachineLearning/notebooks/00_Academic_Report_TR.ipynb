## 2. Çelik Plaka Hata Tespiti – Makine Öğrenme (Bu Proje)

### 2.1 Veri Hattı

Çelik plaka Makine Öğrenme projesi, Veri Madenciliği projesinin ürettiği mühendislik yapılmış veri setini kullanır:

- Girdi dosyası: `Project_2_MachineLearning/data/processed/steel_plates_engineered.csv`
- Kaynak: `Project_1_DataMining` içindeki `03_feature_engineering_EN/TR`
- Özellik sayısı: 27 orijinal + 23 yeni ≈ 50
- Hedef: hata sınıfı (çok sınıflı sınıflandırma)

Veri hattı şu şekilde özetlenebilir:

```
Project_1_DataMining           →   Project_2_MachineLearning
01_data_exploration            →   01_model_training
02_data_preprocessing          →   02_model_evaluation
03_feature_engineering (50 özellik)
          ↓
steel_plates_engineered.csv → modeller ve metrikler
```

### 2.2 Model Eğitimi (01_model_training)

`01_model_training_EN/TR` defterlerinde, mühendislik yapılmış çelik veri seti üzerinde **8 farklı sınıflandırma modeli** eğittik ve karşılaştırdık:

1. **Logistic Regression** – Doğrusal temel model
2. **Decision Tree** – Kural tabanlı, yorumlanabilir model
3. **Random Forest** – Karar ağaçlarından oluşan topluluk
4. **K-Nearest Neighbors (KNN)** – Örnek tabanlı model
5. **Support Vector Machine (SVM)** – Marjine dayalı sınıflandırıcı
6. **Naive Bayes** – Olasılıksal sınıflandırıcı
7. **Neural Network (MLPClassifier)** – Çok katmanlı algılayıcı
8. **Gradient Boosting** – Güçlendirilmiş ağaçlar

**İş akışı:**

- `steel_plates_engineered.csv` dosyasını yükledik
- Özellikleri `X`, hedefi `y` olarak ayırdık
- Sınıf dağılımını korumak için **stratified train/test** bölmesi yaptık
- Ölçeğe duyarlı algoritmalar için (`LR`, `KNN`, `SVM`, `NN`) `StandardScaler` ile ölçekleme uyguladık
- Her modeli makul hiperparametrelerle eğittik
- Modelleri şu metriklerle değerlendirdik:
  - Accuracy
  - Precision, Recall, F1-Skoru
  - Gerekli yerlerde cross-validation
- Tüm sonuçları `model_comparison_results.csv` dosyasında özetledik

**Model kaydetme:**

Eğitim sonunda `models/` klasöründe:

- `*_model.pkl` – Eğitilmiş tüm modeller
- `scaler.pkl` – Eğitilmiş `StandardScaler`
- `label_encoder.pkl` – Sınıf etiket kodlayıcısı
- `model_comparison_results.csv` – Tüm modellerin metrikleri
- `best_model.pkl` – Seçilen en iyi model

Dosyalarını oluşturduk. Böylece Makine Öğrenme projesi **yeniden üretilebilir ve dağıtıma hazır** hale geldi.

### 2.3 Model Değerlendirme (02_model_evaluation)

`02_model_evaluation_EN/TR` defterlerinde, özellikle en iyi modeller (örneğin Random Forest) için daha **derin analiz** yaptık:

- Karışıklık matrisleri (confusion matrix) ürettik
- Her sınıf için precision, recall, F1-skorlarını inceledik
- Hangi hata tiplerinin daha zor tahmin edildiğini analiz ettik
- Sonuçları `figures/` klasöründeki grafikler ile görselleştirdik

Bu değerlendirme adımı, sayısal metrikleri gerçek **hata tespiti bağlamına** bağlar: Hangi hata tiplerinde model güçlü, hangilerinde zayıf, hangi sınıflar karışıyor vb.

## 3. Çelik Plaka Model Optimizasyonu – Ayrı Proje

Optimizasyon aşaması, bağımsız bir proje olarak tasarlandı: `Project_3_Optimization`.

### 3.1 Amaç

Mühendislik yapılmış çelik veri seti üzerinde Random Forest için farklı **hiperparametre arama yöntemlerini** karşılaştırmak:

- **Grid Search** – Önceden tanımlanmış bir parametre ızgarasını tam tarama
- **Random Search** – Parametre aralıklarından rastgele örnekleme

### 3.2 İş Akışı (01_model_optimization)

- `Project_3_Optimization/data/processed/steel_plates_engineered.csv` dosyasını yükledik
- Sınıflara göre stratified train/test bölmesi yaptık
- Gerekli durumlarda `StandardScaler` uyguladık
- Parametre alanlarını tanımladık:
  - Grid Search için: belirli değer listeleri (ör. `n_estimators`, `max_depth`, `min_samples_split`)
  - Random Search için: aralık üzerinden dağılımlar (ör. `randint` ile `n_estimators`)
- Her yöntem için 5-fold cross-validation çalıştırdık
- Şunları ölçtük:
  - En iyi cross-validation skoru
  - Ayrı test kümesi üzerindeki skor
  - Toplam çalışma süresi
- Sonuçları, parametrik aramaları gösteren grafiklerle görselleştirdik

### 3.3 Optimize Edilmiş Modellerin Kaydedilmesi

Optimizasyon sonunda şu dosyalar kaydedildi:

- `random_forest_optimized_grid.pkl` – Grid Search ile bulunan en iyi model
- `random_forest_optimized_random.pkl` – Random Search ile bulunan en iyi model
- `best_optimized_model.pkl` – Genel olarak en iyi optimize edilmiş model
- `optimized_model_comparison.csv` – Yöntem ve skor karşılaştırmaları

Bu yapı, ml-project içindeki “Phase 3: Hyperparameter Tuning” bölümünün, çelik projesi için ayrı bir **Optimizasyon projesi** olarak yeniden tasarlanmış halidir.

## 4. Kalp Damar Hastalığı – Makine Öğrenme (ml-project)

Orijinal **ml-project** içinde Makine Öğrenmesi kısmı, 70.000 kayıt içeren bir **kalp damar hastalığı veri seti** üzerinde gerçekleştirilmiştir. Veri Madenciliği (EDA, ön işleme, özellik mühendisliği) sonrasında çeşitli modeller eğitilip karşılaştırılmıştır.

### 4.1 Veri Hazırlama

- Girdi: `cardio_engineered.csv` (temizlenmiş ve özellik mühendisliği yapılmış veri seti)
- Train/test bölmesi:
  - %80 train, %20 test
  - Hedef değişkene göre stratified bölme (hastalık var / yok)
- Sürekli özellikler için, özellikle mesafe ve marjine dayalı algoritmalarda, `StandardScaler` ile ölçekleme

### 4.2 Eğitilen Modeller

Aşağıdaki 6 sınıflandırma algoritması eğitildi:

1. Lojistik Regresyon
2. Karar Ağacı
3. Rastgele Orman
4. K-En Yakın Komşu (KNN)
5. Destek Vektör Makineleri (SVM)
6. Naive Bayes

Her model şu metriklerle değerlendirildi:

- Accuracy
- Precision
- Recall
- F1-Skoru
- ROC-AUC

**En iyi model**, yaklaşık şu performansla **Random Forest** olmuştur:

- Doğruluk ≈ %73.76
- ROC-AUC ≈ 0.80

### 4.3 Optimizasyon ve Cross-Validation

ml-project akademik raporunda ayrıca:

- Random Forest ve Lojistik Regresyon için **Grid Search**
- SVM için kısmi (kaynak kısıtlı) Grid Search
- 5-fold cross-validation ile model kararlılığı
- Random Forest için özellik önem (feature importance) analizi

adımları gösterilmiştir. Çelik projesinde daha sonra ayrı bir **Optimizasyon projesi** olarak yaptığımız çalışmanın kavramsal karşılığı buradadır.

## Executive Summary

Bu akademik rapor, **iki tam projedeki** Makine Öğrenmesi çalışmalarımızı özetler:

1. **Steel Plates Fault Detection – Machine Learning Project** (`Steel_Fault_Project` / `Project_2_MachineLearning`)
2. Orijinal `ml-project` içindeki **Cardiovascular Disease Prediction – Machine Learning** bölümü

**Steel Fault** projesinde, Data Mining aşamasında üretilen mühendislik yapılmış veri setini (`steel_plates_engineered.csv`, ~50 özellik) kullanarak **8 sınıflandırma modeli** (Logistic Regression, Decision Tree, Random Forest, K‑Nearest Neighbors, Support Vector Machine, Naive Bayes, Neural Network, Gradient Boosting) eğittik. Stratified train/test bölmesi ve gerekli yerlerde ölçekleme yaptık, modelleri Accuracy, Precision, Recall ve F1 ile karşılaştırdık ve tüm modelleri, en iyi modeli ve scaler’ı daha sonra kullanılmak üzere kaydettik. Ayrı bir Optimization projesinde ise, Random Forest için Grid Search ve Random Search ile hiperparametre optimizasyonu gerçekleştirdik.

**Cardiovascular** ml-project’te, temizlenmiş ve özellik mühendisliği yapılmış tıbbi veri seti (`cardio_engineered.csv`) ile çalıştık; birden çok algoritma (Logistic Regression, Decision Tree, Random Forest, KNN, SVM, Naive Bayes) eğittik, bunları Accuracy, F1 ve ROC‑AUC metrikleriyle değerlendirdik ve özellikle Random Forest ve Logistic Regression için Grid Search hiperparametre ayarı yaptık. En iyi model (Random Forest) yaklaşık **%73–74 doğruluk** ve ROC‑AUC ≈ 0.80 elde etti.

Bu iki Makine Öğrenmesi projesi, aynı boru hattının – **train/test bölmesi, ölçekleme, çoklu model karşılaştırması, hiperparametre optimizasyonu ve model kaydetme** – hem **endüstriyel** hata tespiti hem de **tıbbi** hastalık tahmini problemlerine başarıyla uygulanabileceğini göstermektedir.
