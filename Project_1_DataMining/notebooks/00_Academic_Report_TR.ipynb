## 2. Çelik Plaka Hata Tespiti – Veri Madenciliği (Bu Proje)

### 2.1 Amaç ve Veri Seti

**Amaç:** Çelik plaka üzerindeki hataların **geometrisini**, **yoğunluğunu** ve **dokusunu** iyi temsil eden, yüksek kaliteli ve mühendislik yapılmış bir özellik kümesi oluşturmak.

**Bağlam:**

- Proje: `Project_1_DataMining` (`Steel_Fault_Project_v2` içinde)
- Ham dosya: `data/raw/steel_plates_fault.csv`
- Görev: Çelik plaka **hata tiplerini** çok sınıflı sınıflandırma
- Araçlar: Python (pandas, numpy, matplotlib, seaborn), scikit-learn, Jupyter Notebook

Veri seti, çelik plaka görüntülerinden çıkarılmış geometrik ve piksel tabanlı ölçümleri ve her kayda karşılık gelen hata sınıfını içermektedir (örneğin çizik, leke vb.).

### 2.2 Keşifsel Veri Analizi (01_data_exploration)

`01_data_exploration_EN.ipynb` ve `01_data_exploration_TR.ipynb` defterlerinde:

- **Veri yapısını** ve **değişken tiplerini** (sürekli / kategorik özellikler, hedef sınıflar) inceledik
- Geometrik ve piksel tabanlı özelliklerin **dağılımlarını** analiz ettik
- Özellikler ile hata sınıfları arasındaki **korelasyonları** görselleştirdik
- **Veri kalitesini** kontrol ettik:
  - Eksik değerler
  - Yinelenen kayıtlar (duplikeler)
  - Bariz imkânsız değerler (örneğin negatif alan, sıfır genişlik vb.)
- **Sınıf dağılımını** inceleyerek hangi hata tiplerinin az / çok olduğunu belirledik

**Öne çıkan EDA bulguları:**

- Veri setinde çok sayıda geometrik özellik vardır (genişlik, yükseklik, alan, çevre, bounding box ölçüleri vb.)
- Bazı sınıflar daha az sıklıkta görüldüğü için, bu durum sonraki model değerlendirmelerinde önemlidir
- Özellikler farklı ölçeklere sahiptir (örneğin alanlar ile oranlar), bu da modellemeden önce ölçekleme gerektirdiğini gösterir

### 2.3 Ön İşleme (02_data_preprocessing)

`02_data_preprocessing_EN/TR` defterlerinde ham veriyi **analize hazır, temiz bir veri setine** dönüştürdük:

- Varsa **duplikeleri** tespit edip kaldırdık
- **Eksik değerleri** sildik veya uygun yöntemlerle doldurduk
- **Değer aralıklarını** doğruladık:
  - Boyut ve alan gibi özelliklerin pozitif olmasını sağladık
  - Açıkça geçersiz olan kayıtları sildik veya düzelttik
- Sürekli değişkenlerde, özellikle ölçeğe duyarlı algoritmalar için, **ölçekleme** (örneğin `StandardScaler`) uyguladık

Ön işleme sonunda temiz veri setini aşağıdaki dosyalara kaydettik:

- `data/processed/steel_plates_preprocessed.csv`
- Ayrıca tekrar üretilebilir deneyler için: `X_train.npy`, `X_test.npy`, `y_train.npy`, `y_test.npy`

**Ön işleme özeti:**

- Az sayıda geçersiz / tutarsız kayıt kaldırıldı
- Tüm geometrik özellikler fiziksel olarak anlamlı aralıklara getirildi
- Özellik mühendisliği ve model eğitimi için sağlam bir temel oluşturuldu

### 2.4 Özellik Mühendisliği (03_feature_engineering)

Bu Veri Madenciliği projesinin kalbi **özellik mühendisliğidir**. Temiz çelik plaka veri seti üzerinde, hataların **şeklini** ve **yoğunluğunu** daha iyi temsil eden yeni özellikler oluşturduk.

Örnek mühendislik özellikleri:

- **Geometrik / Şekil Özellikleri**
  - En-boy oranı (`width / height`)
  - Bounding box alanı ve alan-çevre oranı
  - Merkez koordinatları (`center_x`, `center_y`)
  - Kompaktlık ve extent
  - Alan, çevre ve bounding box kombinasyonlarından türetilen şekil karmaşıklığı ölçüleri

- **Yoğunluk / Doku Özellikleri**
  - Hata bölgesindeki aydınlık aralığı ve ortalama parlaklık
  - Hata bölgesindeki **piksel yoğunluğu**
  - Boyut–aydınlık etkileşimi (örneğin büyük & parlak, küçük & koyu hatalar)

Son aşamada, orijinal ve mühendislik yapılmış tüm özellikleri tek bir veri setinde birleştirip şu dosyaları kaydettik:

- `data/processed/steel_plates_engineered.csv` – ana mühendislik yapılmış veri seti
- `feature_names.txt` – tüm özellik adları
- `engineered_feature_names.txt` – sadece mühendislik yapılmış özellikler
- `new_feature_names.txt` – yeni eklenen özellikler

Bu veri seti (≈50 özellik: 27 orijinal + 23 yeni), şu projelerde **girdi** olarak kullanıldı:

- `Project_2_MachineLearning` (model eğitimi ve değerlendirme)
- `Project_3_Optimization` (hiperparametre optimizasyonu)

**Sonuç:**

- Hataların geometri ve yoğunluğunu daha bilgilendirici şekilde kodlayan bir özellik kümesi elde edildi
- Ham ölçümlere göre daha güçlü bir temsil sağlandı
- Ham çelik verisinden ML için hazır özelliklere uzanan net ve belgelenmiş bir veri hattı oluşturuldu

## 3. Kalp Damar Hastalığı – Veri Madenciliği (ml-project)

Orijinal **ml-project** içinde Veri Madenciliği kısmı, 70.000 kayıt içeren **tıbbi bir veri setine** odaklanıyordu. Alan farklı olsa da veri madenciliği yaklaşımı aynıdır.

### 3.1 Keşifsel Veri Analizi

Şu adımlar uygulandı:

- Veri setinin şekli, tipleri, eksik değerleri ve duplikeleri kontrol edildi
- Yaş (gün cinsinden), boy, kilo, tansiyon, kolesterol, glukoz, sigara, alkol ve aktivite değişkenlerinin dağılımları incelendi
- Önemli **veri kalitesi sorunları** tespit edildi:
  - Yaşın **gün** cinsinden tutulması
  - Boy, kilo ve tansiyonda **uç değerler**
  - Diastolik tansiyon ≥ sistolik tansiyon olan **imkânsız** kayıtlar
- Hedef değişkenin yaklaşık **dengeli** (%50–50) olduğu doğrulandı

### 3.2 Ön İşleme

Temel adımlar:

- Yaş, **günden yıla** çevrildi
- **Tıbbi bilgiye** dayalı aralıklarla uç değerler silindi
- Fizyolojik olarak **imkânsız** kayıtlar (ör. diastolik ≥ sistolik) kaldırıldı
- Eksik değer ve duplikelerin olmadığı teyit edildi

Son veri seti boyutu: **68.454 kayıt** (verinin %97.79’u korunmuş, %2.21’i silinmiş). Sınıf dengesi korunmuştur.

### 3.3 Özellik Mühendisliği

Toplam **7 tıbbi olarak anlamlı** yeni özellik üretildi:

- BMI ve BMI kategorisi
- Pulse pressure
- MAP (Mean Arterial Pressure)
- Tansiyon kategorisi (AHA sınıfları)
- Yaşam tarzı risk skoru (sigara, alkol, aktivite)
- Yaş grupları

Bu özellikler:

- Kalp damar hastalığı riskine dair alan bilgisini veri setine yansıttı
- Hedef değişkenle güçlü korelasyonlar gösterdi
- ml-project’te model performansını belirgin şekilde artırdı

## 4. Karşılaştırma ve Çıkarımlar (Veri Madenciliği)

Farklı alanlar kullanılmasına rağmen (endüstriyel vs tıbbi), iki projede de aynı **veri madenciliği ilkeleri** uygulanmıştır:

- **Alan:**
  - Çelik proje: Endüstriyel **görüntü / geometri** verisi → **şekil ve yoğunluk** özellikleri
  - Kalp proje: **Klinik & yaşam tarzı** verisi → **tıbbi risk** özellikleri

- **Ön işleme odağı:**
  - Çelik: Geometrik ve yoğunluk özelliklerinin yapısal geçerliliği
  - Kalp: Fizyolojik olarak mümkün aralıklar ve imkânsız vital bulguların silinmesi

- **Özellik mühendisliği:**
  - Çelik: Geometri / doku temelli (aspect ratio, kompaktlık, parlaklık vb.)
  - Kalp: Tıbbi odaklı (BMI, MAP, tansiyon kategorileri, yaşam tarzı riski)

Her iki projede de **Veri Madenciliği temeldir**:

- Veri temiz ve iyi tasarlanmış özelliklerle zenginleştirilmezse, sonraki Makine Öğrenme ve Optimizasyon adımlarının performansı ciddi şekilde düşer
- Veri madenciliği aşaması, **Veri Madenciliği (Veri Madenciliği)** dersinde öğrendiğimiz kavramları, gerçekçi iki farklı probleme uyarlamaktadır

Bu projeler ile **Veri Madenciliği** dersi açısından şu kazanımlara ulaştık:

- Gerçek veri setleri üzerinde sistematik **EDA** yapma

Bu kazanımlar, derste öğrendiğimiz teorinin hem endüstriyel hata tespiti hem de kalp damar hastalığı tahmini gibi gerçekçi problemler üzerinde uygulanabildiğini göstermektedir.


## Executive Summary

Bu akademik rapor, **iki tam projedeki** Veri Madenciliği çalışmalarımızı özetler:

1. **Steel Plates Fault Detection – Data Mining Project** (`Steel_Fault_Project` / `Project_1_DataMining`)
2. Orijinal `ml-project` içindeki **Cardiovascular Disease Prediction – Data Mining** bölümü

**Steel Fault** projesinde, ham `steel_plates_fault.csv` veri setinden başlayarak ayrıntılı **EDA** yaptık, geçersiz ve tutarsız kayıtları temizledik ve yüzey hatalarının **geometrisini** ve **yoğunluğunu/dokusunu** iyi temsil eden yaklaşık **50 özellik** (27 orijinal + 23 yeni) ürettik. Ortaya çıkan `steel_plates_engineered.csv` veri seti, Makine Öğrenme ve Optimizasyon projelerinin temelini oluşturdu.

**Cardiovascular** projesinde, Kaggle’dan alınan 70.000 kayıt içeren tıbbi veri seti ile çalıştık; yaşı gün yerine yıla çevirdik, **uç değerleri** ve **imkânsız değerleri** (ör. diastolik ≥ sistolik) kaldırdık ve **7 tıbbi olarak anlamlı** yeni özellik (BMI, MAP, Pulse Pressure, BP categories, Lifestyle Risk Score, Age Groups) oluşturduk. Sonuçta 68.454 kayıttan oluşan temiz bir veri seti elde ettik ve bu mühendislik yapılmış özellikler ml-project’te model performansını belirgin şekilde artırdı.

Bu iki proje birlikte, **Veri Madenciliği** dersinde öğrendiğimiz EDA, ön işleme ve özellik mühendisliği kavramlarının hem **endüstriyel** (steel faults) hem de **tıbbi** (cardiovascular disease) ortamlarda nasıl uygulanabildiğini göstermektedir.
